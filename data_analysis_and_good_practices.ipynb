# Definição do problema

Este trabalho tem como objetivo analisar como a população alfabetizada se distribui entre os bairros do município do Rio de Janeiro e como essa distribuição evoluiu ao longo dos anos de 2000, 2010 e 2022, com foco na concentração de volumes absolutos da população alfabetizada entre bairros.

- sexo (masculino e feminino)
- estrutura etária (distribuição por faixas de idade)

# Métricas utilizadas

Para avaliar heterogeneidade territorial dos volumes absolutos com apoio em estatística descritiva e análise de concentração de volumes absolutos, serão utilizadas:

- **Totais por bairro** (base absoluta do fenômeno).
- **Participação no total observado por bairro** (`share`), usada como **medida descritiva, não ajustada por população**.
- **Coeficiente de variação (CV)**: mede dispersão relativa entre bairros.
- **Amplitude interquartil (IQR = p75 - p25)**: dispersão robusta, menos sensível a extremos.
- **Razão p90/p10**: contraste entre bairros no topo e na base da distribuição.
- **Índice de Gini** (0 a 1): medida clássica de desigualdade na distribuição dos volumes absolutos.
- **Comparações temporais (2000, 2010, 2022)** sobre base harmonizada.

Essas métricas combinam leitura intuitiva (totais e participações descritivas) com medidas robustas e consagradas de desigualdade de volumes (CV, IQR, p90/p10 e Gini).

```python
#13

from pathlib import Path
import urllib.request

# Leitura dos arquivos via GitHub Raw (prioritario para Colab)
BASE_RAW = "https://raw.githubusercontent.com/vdms/analise_de_dados_e_boas_praticas/main/data"
BASE_LOCAL = Path("data")

def carregar_csv(nome_arquivo):
    url = f"{BASE_RAW}/{nome_arquivo}"
    try:
        return pd.read_csv(url, sep=";", encoding="utf-8")
    except Exception:
        caminho_local = BASE_LOCAL / nome_arquivo
        return pd.read_csv(caminho_local, sep=";", encoding="utf-8")

df_2000 = carregar_csv("495_2000.csv")
df_2010 = carregar_csv("495_2010.csv")
df_2022 = carregar_csv("495_2022.csv")

# Leiame (opcional)
leiame = ""
leiame_url = f"{BASE_RAW}/leiame.txt"
try:
    with urllib.request.urlopen(leiame_url) as f:
        leiame = f.read().decode("utf-8")
except Exception:
    leiame_local = BASE_LOCAL / "leiame.txt"
    if leiame_local.exists():
        leiame = leiame_local.read_text(encoding="utf-8")

print("2000:", df_2000.shape)
print("2010:", df_2010.shape)
print("2022:", df_2022.shape)
```

```python
#14
# Adicionar coluna de ano para facilitar comparações futuras

df_2000["ano"] = 2000
df_2010["ano"] = 2010
df_2022["ano"] = 2022
```

```python
#15
# Limpar e padronizar os nomes das colunas para facilitar o trabalho futuro
import re
import unicodedata

def normalizar_coluna(col):
    col = unicodedata.normalize("NFKD", col).encode("ascii", "ignore").decode("ascii")
    col = col.strip().lower()
    col = re.sub(r"[^a-z0-9]+", "_", col)
    col = re.sub(r"_+", "_", col).strip("_")
    return col

def limpar_colunas(df):
    df = df.copy()
    df.columns = [normalizar_coluna(c) for c in df.columns]
    return df

df_2000 = limpar_colunas(df_2000)
df_2010 = limpar_colunas(df_2010)
df_2022 = limpar_colunas(df_2022)
```

# Comparabilidade territorial entre os anos

Antes de comparar os anos, é necessário verificar se a cobertura territorial é compatível entre as bases. Para isso, utiliza-se `codbnum` como chave territorial principal, por ser mais estável do que o nome do bairro.

Essa etapa permite identificar quais unidades territoriais estão presentes simultaneamente em 2000, 2010 e 2022, definindo assim o conjunto de bairros que poderá sustentar a comparação temporal principal.


```python
#31
# Verificar duplicidade e comparabilidade territorial entre os anos

# Usamos codbnum como chave territorial principal (mais estável que nome de bairro)
chave_geo = "codbnum"
bases = {2000: df_2000, 2010: df_2010, 2022: df_2022}
codigos_por_ano = {}

for ano, df in bases.items():
    print(f"\nAno {ano}")
    print("Número de linhas:", len(df))
    print("Bairros únicos:", df["bairro"].nunique())
    print("Códigos territoriais únicos:", df[chave_geo].nunique())
    print("Linhas duplicadas por codbnum:", df.duplicated(subset=[chave_geo]).sum())
    codigos_por_ano[ano] = set(df[chave_geo].dropna().astype(int))

codigos_comuns = set.intersection(*codigos_por_ano.values())
print("\nCódigos territoriais em comum nos três anos:", len(codigos_comuns))

for ano in [2000, 2010, 2022]:
    exclusivos = codigos_por_ano[ano] - codigos_comuns
    print(f"Códigos exclusivos de {ano}: {len(exclusivos)}")

# Bases harmonizadas para comparação temporal principal
df_2000_comp = df_2000[df_2000[chave_geo].isin(codigos_comuns)].copy()
df_2010_comp = df_2010[df_2010[chave_geo].isin(codigos_comuns)].copy()
df_2022_comp = df_2022[df_2022[chave_geo].isin(codigos_comuns)].copy()

print("\nDimensões harmonizadas (painel balanceado por codbnum):")
print("2000:", df_2000_comp.shape)
print("2010:", df_2010_comp.shape)
print("2022:", df_2022_comp.shape)

# Auditoria das exclusões territoriais em 2010 e 2022

def colunas_alfabetizacao_sexo(df):
    return [c for c in df.columns if "alfabet" in c and "sexo" in c]


def tabela_exclusoes_territoriais(df, ano, codigos_excluidos):
    cols_alf = colunas_alfabetizacao_sexo(df)
    base_exc = df[df[chave_geo].isin(codigos_excluidos)].copy()

    if base_exc.empty:
        return pd.DataFrame(columns=["ano", "bairro", "codbnum", "total_alfabetizados_ano"])

    base_exc["total_alfabetizados_ano"] = base_exc[cols_alf].sum(axis=1, skipna=True)
    base_exc[chave_geo] = base_exc[chave_geo].astype(int)
    base_exc["ano"] = ano

    return (
        base_exc[["ano", "bairro", "codbnum", "total_alfabetizados_ano"]]
        .sort_values(["total_alfabetizados_ano", "bairro"], ascending=[False, True])
        .reset_index(drop=True)
    )

codigos_excluidos_2010 = sorted(codigos_por_ano[2010] - codigos_comuns)
codigos_excluidos_2022 = sorted(codigos_por_ano[2022] - codigos_comuns)

auditoria_exclusoes_territoriais = pd.concat([
    tabela_exclusoes_territoriais(df_2010, 2010, codigos_excluidos_2010),
    tabela_exclusoes_territoriais(df_2022, 2022, codigos_excluidos_2022),
], ignore_index=True)

print("\nBairros/códigos fora da interseção (2010 e 2022):")
display(auditoria_exclusoes_territoriais)

resumo_exclusoes_territoriais = []
for ano in [2010, 2022]:
    cols_alf_ano = colunas_alfabetizacao_sexo(bases[ano])
    total_ano = bases[ano][cols_alf_ano].sum(axis=1, skipna=True).sum()
    total_excluido = auditoria_exclusoes_territoriais.loc[
        auditoria_exclusoes_territoriais["ano"] == ano, "total_alfabetizados_ano"
    ].sum()

    resumo_exclusoes_territoriais.append({
        "ano": ano,
        "n_bairros_excluidos": int((auditoria_exclusoes_territoriais["ano"] == ano).sum()),
        "total_alfabetizados_excluidos": float(total_excluido),
        "total_alfabetizados_ano": float(total_ano),
        "peso_exclusoes_%": (100 * total_excluido / total_ano) if total_ano > 0 else np.nan,
    })

resumo_exclusoes_territoriais = pd.DataFrame(resumo_exclusoes_territoriais)
print("Resumo de impacto das exclusões territoriais:")
display(resumo_exclusoes_territoriais.round(3))

impacto_max = resumo_exclusoes_territoriais["peso_exclusoes_%"].max()
if impacto_max < 2.0:
    parecer_exclusoes_territoriais = (
        "As exclusões auditadas parecem não alterar substantivamente a leitura geral da distribuição territorial."
    )
else:
    parecer_exclusoes_territoriais = (
        "As exclusões auditadas podem alterar de forma relevante a leitura geral e devem ser consideradas na interpretação final."
    )

print("Parecer metodológico:", parecer_exclusoes_territoriais)

```

A verificação confirma ausência de duplicidade territorial quando usamos `codbnum`, preservando a unidade de análise da base.

Para a comparação temporal principal, foi adotada a interseção de códigos territoriais comuns aos três anos (painel balanceado), por ser uma estratégia que reduz variação de cobertura territorial entre períodos. Essa escolha foi usada para manter consistência interna da comparação neste estudo.

## Política adotada para nulos harmonizados

Critério formal de exclusão no painel comparativo:

1. o território precisa estar na interseção 2000-2010-2022;
2. o território apresenta pelo menos um `NaN` em coluna harmonizada usada no total comparativo;
3. o número de territórios afetados é pequeno (`<= 10`), permitindo exclusão controlada sem comprometer a cobertura global.

Regra operacional adotada: os códigos territoriais afetados são excluídos em todos os anos da base comparativa.

Justificativa metodológica: nesses casos, a imputação exigiria reconstrução de contagens em múltiplas células etária/sexo, com risco de alterar totais por bairro e métricas de desigualdade. A exclusão simétrica foi adotada para manter a mesma lista de territórios em todos os anos e preservar rastreabilidade do pré-processamento. Nesta versão, esse procedimento não foi comparado com alternativas como soma parcial por `skipna` ou imputação simples.


```python
#41

# Aplicar politica explicita para nulos harmonizados antes da base final

def codigos_afetados_painel(df_auditoria):
    return set(
        df_auditoria.loc[df_auditoria["entra_na_base_harmonizada"], "codbnum"]
        .dropna()
        .astype(int)
    )

casos_painel_nulos = auditoria_nulos_harmonizados_df[auditoria_nulos_harmonizados_df["entra_na_base_harmonizada"]].copy()
limite_casos_pequenos = 10

shape_antes = {
    2000: df_2000_comp.shape[0],
    2010: df_2010_comp.shape[0],
    2022: df_2022_comp.shape[0],
}

if len(casos_painel_nulos) <= limite_casos_pequenos:
    politica_nulos_harmonizados = "excluir_codigos_em_todos_os_anos_para_preservar_painel_balanceado"
    codigos_excluir_painel = codigos_afetados_painel(casos_painel_nulos)

    df_2000_comp = df_2000_comp[~df_2000_comp["codbnum"].isin(codigos_excluir_painel)].copy()
    df_2010_comp = df_2010_comp[~df_2010_comp["codbnum"].isin(codigos_excluir_painel)].copy()
    df_2022_comp = df_2022_comp[~df_2022_comp["codbnum"].isin(codigos_excluir_painel)].copy()
else:
    politica_nulos_harmonizados = "manter_linhas_sem_exclusao"
    codigos_excluir_painel = set()

shape_depois = {
    2000: df_2000_comp.shape[0],
    2010: df_2010_comp.shape[0],
    2022: df_2022_comp.shape[0],
}

resumo_politica_nulos = pd.DataFrame({
    "ano": [2000, 2010, 2022],
    "linhas_antes": [shape_antes[2000], shape_antes[2010], shape_antes[2022]],
    "linhas_depois": [shape_depois[2000], shape_depois[2010], shape_depois[2022]],
    "linhas_excluidas": [
        shape_antes[2000] - shape_depois[2000],
        shape_antes[2010] - shape_depois[2010],
        shape_antes[2022] - shape_depois[2022],
    ],
})

print("Politica aplicada para nulos harmonizados:", politica_nulos_harmonizados)
print("Codigos excluidos para manter painel balanceado:", len(codigos_excluir_painel))

if shape_antes[2000] > 0:
    perc_excluidos = 100 * len(codigos_excluir_painel) / shape_antes[2000]
    print(f"Territorios excluidos no painel: {len(codigos_excluir_painel)} de {shape_antes[2000]} ({perc_excluidos:.2f}%).")

display(resumo_politica_nulos)

painel_balanceado_pos_nulos = len(set(shape_depois.values())) == 1
print("Painel balanceado apos tratamento de nulos:", painel_balanceado_pos_nulos)

if not painel_balanceado_pos_nulos:
    raise ValueError("A base comparativa deixou de ser painel balanceado apos o tratamento de nulos.")

if len(casos_painel_nulos) > 0:
    territorios_excluidos = (
        casos_painel_nulos[["codbnum", "bairro"]]
        .drop_duplicates()
        .sort_values("codbnum")
        .reset_index(drop=True)
    )
    print("Territorios excluidos por nulos harmonizados:")
    display(territorios_excluidos)

    display(
        casos_painel_nulos[["ano", "bairro", "codbnum", "colunas_harmonizadas_com_nulo", "n_campos_ausentes_linha"]]
        .sort_values(["ano", "n_campos_ausentes_linha"], ascending=[True, False])
    )
```

Com essa regra, o efeito dos valores ausentes sobre os totais permanece rastreável: os casos afetados são identificados, sua entrada no painel comparativo é verificada e o pré-processamento aplicado é explicitado antes da construção da base harmonizada final.

```python
#45a
def construir_visao_substantiva(df):
    base = df[["bairro", "codbnum", "ano"]].copy()

    col_m = [c for c in df.columns if c.startswith("alf_m_")]
    col_f = [c for c in df.columns if c.startswith("alf_f_")]

    base["total_m"] = df[col_m].sum(axis=1)
    base["total_f"] = df[col_f].sum(axis=1)
    base["total_alfabetizados"] = base["total_m"] + base["total_f"]

    base["jovens"] = df[[c for c in df.columns if any(x in c for x in ["15_19", "20_24", "25_29"])]].sum(axis=1)
    base["adultos"] = df[[c for c in df.columns if any(x in c for x in ["30_34", "35_39", "40_44", "45_49", "50_54", "55_59"])]].sum(axis=1)
    base["idosos"] = df[[c for c in df.columns if any(x in c for x in ["60_64", "65_69", "70_79", "80_mais"])]].sum(axis=1)

    return base

def aplicar_minmax(df, colunas):
    out = df.copy()
    for c in colunas:
        c_min = out[c].min()
        c_max = out[c].max()
        if pd.isna(c_min) or pd.isna(c_max) or c_max == c_min:
            out[f"{c}_norm"] = 0.0
        else:
            out[f"{c}_norm"] = (out[c] - c_min) / (c_max - c_min)
    return out

visao_substantiva = pd.concat([
    construir_visao_substantiva(base_2000_h),
    construir_visao_substantiva(base_2010_h),
    construir_visao_substantiva(base_2022_h),
], ignore_index=True)

variaveis_norm = [
    "total_alfabetizados",
    "total_f",
    "total_m",
    "jovens",
    "adultos",
    "idosos",
]

visao_normalizada_minmax = aplicar_minmax(visao_substantiva, variaveis_norm)

display(visao_normalizada_minmax[[
    "bairro", "codbnum", "ano",
    "total_alfabetizados_norm",
    "total_f_norm",
    "total_m_norm",
    "jovens_norm",
    "adultos_norm",
    "idosos_norm",
]].head())
```

# Métricas de desigualdade territorial na base harmonizada

Concluídas as etapas de preparação de dados, a análise principal segue com as variáveis harmonizadas em escala original para preservar interpretabilidade territorial e comparabilidade estrutural entre os anos.


```python
#47
# Cálculo de métricas de desigualdade territorial (base harmonizada etariamente)

def gini(x):
    x = np.asarray(x, dtype=float)
    x = x[np.isfinite(x)]
    if len(x) == 0:
        return np.nan
    if np.any(x < 0):
        return np.nan
    if np.all(x == 0):
        return 0.0
    x_sorted = np.sort(x)
    n = len(x_sorted)
    cumx = np.cumsum(x_sorted)
    return (n + 1 - 2 * (cumx.sum() / cumx[-1])) / n

def resumo_desigualdade(df, ano):
    cols = [c for c in df.columns if c.startswith("alf_")]
    if not cols:
        raise ValueError(f"Nenhuma coluna 'alf_' encontrada na base de {ano}.")

    total_bairro = df[cols].sum(axis=1)
    q10, q25, q50, q75, q90 = np.quantile(total_bairro, [0.10, 0.25, 0.50, 0.75, 0.90])
    media = total_bairro.mean()
    cv = total_bairro.std(ddof=0) / media if media != 0 else np.nan
    return {
        "ano": ano,
        "n_bairros": int(total_bairro.shape[0]),
        "total_alfabetizados": float(total_bairro.sum()),
        "media_bairro": float(media),
        "mediana_bairro": float(q50),
        "iqr": float(q75 - q25),
        "p90_p10": float(q90 / q10) if q10 > 0 else np.nan,
        "cv": float(cv),
        "gini": float(gini(total_bairro)),
    }

metricas = pd.DataFrame([
    resumo_desigualdade(base_2000_h, 2000),
    resumo_desigualdade(base_2010_h, 2010),
    resumo_desigualdade(base_2022_h, 2022),
])

display(metricas.round(4))
```

<!-- #48 -->
Após a harmonização, as três bases passam a ter a mesma estrutura analítica, com as mesmas faixas etárias comparáveis e a mesma separação por sexo. Esse passo assegura comparabilidade estrutural das variáveis selecionadas entre os anos.

Essa harmonização não implica comparabilidade plena de todos os aspectos substantivos dos censos, que podem incluir diferenças de cobertura territorial, dinâmica urbana e procedimentos estatísticos. No escopo do notebook, a harmonização sustenta a comparação descritiva das métricas calculadas e da composição por sexo e estrutura etária.

# Análise da distribuição territorial dos totais de população alfabetizada

Com a base harmonizada, torna-se possível analisar como os totais de população alfabetizada se distribuem entre os bairros e como essa distribuição evolui ao longo do tempo.

O foco desta etapa é identificar padrões de concentração de volumes absolutos e dispersão da população alfabetizada, utilizando métricas que permitam comparar os anos de 2000, 2010 e 2022 de forma consistente.

A análise será conduzida a partir de medidas de variabilidade e de desigualdade na distribuição territorial dos totais, complementadas por visualizações que facilitem a interpretação dos resultados.



```python
#50
def calcular_total(df):
    colunas_alf = [c for c in df.columns if c.startswith("alf_")]
    df["total_alfabetizados"] = df[colunas_alf].sum(axis=1)
    return df

base_2000_h = calcular_total(base_2000_h)
base_2010_h = calcular_total(base_2010_h)
base_2022_h = calcular_total(base_2022_h)
```

Essa variável agregada permite comparar diretamente o volume de população alfabetizada entre bairros, servindo como base para o cálculo das métricas de heterogeneidade e de concentração de volumes absolutos.


```python
#52
def resumo_estatistico(df, ano):
    print(f"\nAno {ano}")
    display(df["total_alfabetizados"].describe())

resumo_estatistico(base_2000_h, 2000)
resumo_estatistico(base_2010_h, 2010)
resumo_estatistico(base_2022_h, 2022)
```

As estatísticas descritivas permitem uma primeira avaliação da distribuição da população alfabetizada entre os bairros.

A comparação entre média, mediana, valores mínimos e máximos fornece indícios iniciais de heterogeneidade territorial nos volumes absolutos. Em particular, diferenças relevantes entre média e mediana e a presença de valores máximos muito elevados podem indicar concentração da população em um número reduzido de bairros.


```python
#54
def calcular_metricas(df):
    s = df["total_alfabetizados"]
    
    return {
        "media": s.mean(),
        "mediana": s.median(),
        "desvio_padrao": s.std(),
        "coef_variacao": s.std() / s.mean(),
        "p90_p10": s.quantile(0.9) / s.quantile(0.1),
        "iqr": s.quantile(0.75) - s.quantile(0.25)
    }

metricas = pd.DataFrame({
    "2000": calcular_metricas(base_2000_h),
    "2010": calcular_metricas(base_2010_h),
    "2022": calcular_metricas(base_2022_h)
}).T

display(metricas)
```

A leitura da tabela de métricas descreve desigualdade nos volumes absolutos por bairro ao longo do período. Em conjunto, CV, p90/p10 e Gini permanecem elevados, enquanto o IQR não evolui de forma monotônica entre 2000, 2010 e 2022.

Esse padrão é compatível com heterogeneidade persistente dos totais observados entre bairros. Como os indicadores são calculados sobre contagens absolutas, efeitos de escala associados ao porte populacional dos territórios influenciam os resultados e limitam interpretações em termos relativos.

Para manter coerência com a análise já desenvolvida e evitar mistura de estruturas entre anos, a matriz de correlação é construída com a base harmonizada de **2022** como ano de referência.

As variáveis selecionadas combinam medidas de volume (`total_alfabetizados`, `total_f`, `total_m`) e de composição (`prop_f`, `prop_jovens`, `prop_idosos`), permitindo verificar como essas dimensões se articulam no recorte territorial por bairro.


```python
#57b-prep
# Matriz de correlacao (Pearson) no ano de referencia 2022

base_corr_2022 = base_2022_h.copy()

col_m = [c for c in base_corr_2022.columns if c.startswith("alf_m_")]
col_f = [c for c in base_corr_2022.columns if c.startswith("alf_f_")]

base_corr_2022["total_m"] = base_corr_2022[col_m].sum(axis=1)
base_corr_2022["total_f"] = base_corr_2022[col_f].sum(axis=1)

base_corr_2022["jovens"] = base_corr_2022[[c for c in base_corr_2022.columns if any(x in c for x in ["15_19", "20_24", "25_29"]) ]].sum(axis=1)
base_corr_2022["idosos"] = base_corr_2022[[c for c in base_corr_2022.columns if any(x in c for x in ["60_64", "65_69", "70_79", "80_mais"]) ]].sum(axis=1)

den = base_corr_2022["total_alfabetizados"].replace(0, np.nan)
base_corr_2022["prop_f"] = base_corr_2022["total_f"] / den
base_corr_2022["prop_jovens"] = base_corr_2022["jovens"] / den
base_corr_2022["prop_idosos"] = base_corr_2022["idosos"] / den

vars_corr = [
    "total_alfabetizados",
    "total_f",
    "total_m",
    "prop_f",
    "prop_jovens",
    "prop_idosos",
]

corr_2022 = base_corr_2022[vars_corr].corr(method="pearson")

rotulos = {
    "total_alfabetizados": "total_alfabetizados",
    "total_f": "total_f",
    "total_m": "total_m",
    "prop_f": "prop_f",
    "prop_jovens": "prop_jovens",
    "prop_idosos": "prop_idosos",
}
corr_plot = corr_2022.rename(index=rotulos, columns=rotulos)
```

```python
#57b-plot
import seaborn as sns

fig, ax = plt.subplots(figsize=FIG_SIZE_HEATMAP)

sns.heatmap(
    corr_plot,
    cmap="coolwarm",
    vmin=-1,
    vmax=1,
    annot=True,
    fmt=".2f",
    linewidths=0,
    cbar_kws={"label": "Correlação de Pearson"},
    ax=ax,
)

ax.set_title("Figure 3.3 — Matriz de correlação (ano de referência: 2022)")
plt.tight_layout()
plt.show()
```

## Hipótese alternativa

Embora os resultados sugiram aumento da concentração de volumes absolutos de população alfabetizada em determinados bairros, é importante considerar explicações alternativas para esse padrão.

Uma possibilidade é que o crescimento observado esteja associado ao aumento absoluto da população em bairros mais densos, e não necessariamente a um processo de intensificação da desigualdade territorial. Nesse caso, a elevação dos valores máximos refletiria expansão populacional concentrada, e não redistribuição desigual entre territórios. Isso implicaria um padrão de crescimento concentrado, mas não necessariamente aumento da desigualdade relativa entre bairros.

Outra hipótese é que mudanças na estrutura territorial ou nos critérios de agregação dos dados entre os censos tenham afetado a comparabilidade entre os anos, contribuindo para parte da variação observada.

Por fim, diferenças metodológicas entre os censos, incluindo ajustes por imputação estatística, podem introduzir variações que não refletem mudanças reais na distribuição da população.

Essas hipóteses alternativas não invalidam os resultados encontrados, mas indicam a necessidade de interpretar os padrões observados com cautela.

## Limitações da análise

Os resultados apresentados devem ser interpretados considerando algumas limitações relevantes da base de dados e da abordagem adotada.

Primeiro, a análise depende de uma harmonização das faixas etárias entre os anos, o que implica perda de granularidade e pode ocultar variações internas mais específicas.

Por fim, os dados resultam de processos estatísticos que incluem imputações, o que pode gerar pequenas diferenças em relação aos totais oficiais. Também não foram realizados testes formais de robustez ou análises de sensibilidade para estratégias alternativas de painel e tratamento de nulos.

Essas limitações delimitam o alcance das conclusões, que devem ser lidas como evidência descritiva condicional às escolhas metodológicas adotadas.

Após analisar a concentração de volumes absolutos dos totais de população alfabetizada de forma agregada, esta seção investiga como esse padrão se manifesta quando consideradas as dimensões de sexo e faixa etária.

O objetivo é verificar se a heterogeneidade de volumes absolutos observada se mantém quando desagregamos os dados, além de identificar possíveis mudanças na composição etária da população alfabetizada ao longo do tempo.


```python
#70
def total_por_sexo(df):
    col_m = [c for c in df.columns if c.startswith("alf_m_")]
    col_f = [c for c in df.columns if c.startswith("alf_f_")]

    df["total_m"] = df[col_m].sum(axis=1)
    df["total_f"] = df[col_f].sum(axis=1)

    return df

base_2000_h = total_por_sexo(base_2000_h)
base_2010_h = total_por_sexo(base_2010_h)
base_2022_h = total_por_sexo(base_2022_h)
```

```python
#73
# Estatísticas por ano para prop_f e correlação com total_alfabetizados

bases_por_ano = {
    2000: base_2000_h,
    2010: base_2010_h,
    2022: base_2022_h,
}

resumo_sexo = []
corr_sexo = []

for ano, df in bases_por_ano.items():
    serie = df["prop_f"].dropna()
    q1 = serie.quantile(0.25)
    q3 = serie.quantile(0.75)

    resumo_sexo.append({
        "ano": ano,
        "media_prop_f": serie.mean(),
        "mediana_prop_f": serie.median(),
        "iqr_prop_f": q3 - q1,
        "min_prop_f": serie.min(),
        "max_prop_f": serie.max(),
    })

    corr = df[["total_alfabetizados", "prop_f"]].corr(method="pearson").iloc[0, 1]
    corr_sexo.append({"ano": ano, "corr_total_prop_f": corr})

resumo_prop_f = pd.DataFrame(resumo_sexo).sort_values("ano")
corr_prop_f = pd.DataFrame(corr_sexo).sort_values("ano")

resumo_prop_f_exib = resumo_prop_f.copy()
for col in ["media_prop_f", "mediana_prop_f", "iqr_prop_f", "min_prop_f", "max_prop_f"]:
    resumo_prop_f_exib[col] = (resumo_prop_f_exib[col] * 100).round(2)

resumo_prop_f_exib = resumo_prop_f_exib.rename(columns={
    "media_prop_f": "media_prop_f_%",
    "mediana_prop_f": "mediana_prop_f_%",
    "iqr_prop_f": "iqr_prop_f_pp",
    "min_prop_f": "min_prop_f_%",
    "max_prop_f": "max_prop_f_%",
})

print("Resumo de prop_f por ano")
display(resumo_prop_f_exib)

print("Correlação de Pearson entre total_alfabetizados e prop_f por ano")
display(corr_prop_f.assign(corr_total_prop_f=corr_prop_f["corr_total_prop_f"].round(3)))

```

```python
#75
def estrutura_etaria(df):
    df["jovens"] = df[[c for c in df.columns if any(x in c for x in ["15_19", "20_24", "25_29"]) ]].sum(axis=1)
    df["adultos"] = df[[c for c in df.columns if any(x in c for x in ["30_34", "35_39", "40_44", "45_49", "50_54", "55_59"]) ]].sum(axis=1)
    df["idosos"] = df[[c for c in df.columns if any(x in c for x in ["60_64", "65_69", "70_79", "80_mais"]) ]].sum(axis=1)

    return df

base_2000_h = estrutura_etaria(base_2000_h)
base_2010_h = estrutura_etaria(base_2010_h)
base_2022_h = estrutura_etaria(base_2022_h)
```

```python
#78-prep
# Indicadores combinados de envelhecimento relativo: medianas, IQR e diferença 2000-2022

bases_etaria = {
    2000: base_2000_h,
    2010: base_2010_h,
    2022: base_2022_h,
}

linhas_etaria = []
for ano, df in bases_etaria.items():
    jovens = df["prop_jovens"].dropna()
    idosos = df["prop_idosos"].dropna()

    linhas_etaria.append({
        "ano": str(ano),
        "mediana_prop_jovens": jovens.median(),
        "mediana_prop_idosos": idosos.median(),
        "iqr_prop_jovens": jovens.quantile(0.75) - jovens.quantile(0.25),
        "iqr_prop_idosos": idosos.quantile(0.75) - idosos.quantile(0.25),
    })

resumo_etario = pd.DataFrame(linhas_etaria).sort_values("ano")

base_2000 = resumo_etario.loc[resumo_etario["ano"] == "2000"].iloc[0]
base_2022 = resumo_etario.loc[resumo_etario["ano"] == "2022"].iloc[0]

linha_diff = {
    "ano": "dif_2000_2022",
    "mediana_prop_jovens": base_2022["mediana_prop_jovens"] - base_2000["mediana_prop_jovens"],
    "mediana_prop_idosos": base_2022["mediana_prop_idosos"] - base_2000["mediana_prop_idosos"],
    "iqr_prop_jovens": base_2022["iqr_prop_jovens"] - base_2000["iqr_prop_jovens"],
    "iqr_prop_idosos": base_2022["iqr_prop_idosos"] - base_2000["iqr_prop_idosos"],
}

tabela_etaria = pd.concat([resumo_etario, pd.DataFrame([linha_diff])], ignore_index=True)

tabela_etaria_exib = tabela_etaria.copy()
for col in ["mediana_prop_jovens", "mediana_prop_idosos", "iqr_prop_jovens", "iqr_prop_idosos"]:
    tabela_etaria_exib[col] = (tabela_etaria_exib[col] * 100).round(2)

tabela_etaria_exib = tabela_etaria_exib.rename(columns={
    "mediana_prop_jovens": "mediana_prop_jovens_%",
    "mediana_prop_idosos": "mediana_prop_idosos_%",
    "iqr_prop_jovens": "iqr_prop_jovens_pp",
    "iqr_prop_idosos": "iqr_prop_idosos_pp",
})

print("Estrutura etária por ano e diferença 2000-2022")
display(tabela_etaria_exib)

medianas_plot = resumo_etario.copy()
medianas_plot["ano"] = medianas_plot["ano"].astype(int)
```
