# Definição do problema

Este trabalho tem como objetivo analisar a distribuição espacial e temporal da população alfabetizada por sexo e faixa etária nos bairros do município do Rio de Janeiro, com base em dados agregados do IBGE por setor censitário para os anos 2000, 2010 e 2022.

Mais especificamente, a proposta é responder à seguinte pergunta:

> Como a composição da população alfabetizada mudou entre 2000, 2010 e 2022 nos bairros do Rio de Janeiro, considerando diferenças por sexo e grupos etários comparáveis?

A escolha desse problema é adequada ao conjunto de dados por quatro razões:

1. Os arquivos já estão organizados com recortes territoriais e demográficos.
2. O tema permite uma análise descritiva clara, sem exigir modelagem preditiva.
3. Há possibilidade de comparar anos distintos, o que enriquece a narrativa analítica.
4. O problema exige cuidado metodológico, pois os anos não são totalmente comparáveis, o que cria espaço para uma discussão acadêmica relevante sobre limitações, padronização e interpretação.

Este não é um problema de machine learning. Trata-se de um estudo de análise exploratória de dados com pré-processamento, cujo foco é compreender a estrutura da base, identificar padrões, avaliar diferenças entre anos e construir uma leitura consistente e defensável dos dados.

# Recorte analítico

Como os arquivos trazem contagens de pessoas alfabetizadas por bairro, sexo e faixa etária, o recorte adotado será intencionalmente simples e coerente:

Unidade de análise: bairro
Dimensão temporal: anos de 2000, 2010 e 2022
Fenômeno observado: população alfabetizada
Dimensões de comparação: sexo e faixa etária

Para manter a comparabilidade entre os três anos, **a análise principal será concentrada nas faixas etárias presentes em todos os arquivos**.

Esse ponto é essencial, pois a base de 2022 não inclui os grupos de 5 a 9 anos e 10 a 14 anos, e também reorganiza a parte final da distribuição etária, reunindo 70 a 79 anos em uma única faixa. Além disso, o Data.Rio compartilha um "leiame" que informa que, no conjunto de 2022, as faixas para pessoas alfabetizadas começam em 15 anos e terminam em 80 anos ou mais. O "leiame" também informa que a soma dos setores pode diferir da população oficial por conta de imputações estatísticas aplicadas pelo IBGE.

Diante disso, a análise será construída em dois níveis:

## Nível 1 — análise comparável entre 2000, 2010 e 2022

* usar apenas as faixas etárias comuns;
* focar em tendências temporais seguras;
* evitar conclusões baseadas em categorias incompatíveis.

## Nível 2 — análise complementar por ano

* explorar peculiaridades de cada base;
* registrar diferenças metodológicas;
* mostrar o que pode e o que não pode ser comparado diretamente.

Esse recorte fortalece o trabalho, porque mostra rigor metodológico e evita interpretações artificiais.
