# Definição do problema

Este trabalho tem como objetivo analisar como a população alfabetizada se distribui entre os bairros do município do Rio de Janeiro e como essa distribuição evoluiu ao longo dos anos de 2000, 2010 e 2022, com foco na identificação de padrões de desigualdade territorial.

Além disso, a análise será complementada por duas dimensões demográficas:

- sexo (masculino e feminino)
- estrutura etária (distribuição por faixas de idade)

A pergunta central que orienta o estudo é:

> Como a desigualdade na distribuição da população alfabetizada entre bairros evoluiu ao longo do tempo, e como essa desigualdade se manifesta quando observamos diferenças por sexo e estrutura etária?

# Justificativa

A escolha desse problema permite ir além de uma descrição simples dos dados, introduzindo uma perspectiva analítica baseada em distribuição e desigualdade territorial.

Essa abordagem é adequada ao dataset porque:

- os dados estão organizados por bairro, o que permite comparações espaciais diretas;
- a variável de interesse (população alfabetizada) está disponível em forma de contagem, possibilitando análises de concentração e dispersão;
- a presença de diferentes anos permite observar mudanças ao longo do tempo;
- as dimensões de sexo e idade enriquecem a análise, permitindo identificar padrões demográficos dentro da desigualdade observada.

Essa combinação torna possível construir uma análise exploratória consistente, com interpretações claras e bem fundamentadas.

# Hipóteses exploratórias

Como ponto de partida para a análise, são consideradas as seguintes hipóteses:

1. A distribuição da população alfabetizada entre bairros é desigual, com concentração em determinados territórios.
2. Essa desigualdade se mantém ou se intensifica ao longo do tempo.
3. A estrutura etária da população alfabetizada indica um possível processo de envelhecimento.
4. Existem diferenças entre sexos na composição da população alfabetizada, que podem variar entre bairros.

Essas hipóteses não são assumidas como verdadeiras, mas servem como guia para a exploração dos dados.

# Estratégia analítica

A análise será conduzida em quatro níveis complementares:

## 1. Distribuição territorial
- avaliar como a população alfabetizada se distribui entre bairros;
- identificar concentração, dispersão e possíveis desigualdades.

## 2. Evolução temporal
- comparar os anos de 2000, 2010 e 2022;
- verificar mudanças na distribuição.

## 3. Dimensão de sexo
- analisar a participação relativa de homens e mulheres;
- identificar variações entre bairros.

## 4. Estrutura etária
- avaliar a composição por idade;
- investigar sinais de envelhecimento da população alfabetizada.

Essa abordagem permite uma análise progressiva, partindo do geral (território) para o específico (sexo e idade).

# Cuidados metodológicos

A comparação entre os anos exige atenção a diferenças estruturais nos dados:

- as faixas etárias não são totalmente consistentes entre os anos;
- no dataset de 2022, as faixas começam em 15 anos;
- no dataset de 2022, há agregações diferentes nas idades mais avançadas;
- os totais podem não coincidir exatamente com valores oficiais devido a imputações estatísticas realizadas pelo IBGE.

Diante disso, serão adotadas as seguintes estratégias:

- utilizar apenas faixas etárias comparáveis na análise principal;
- evitar comparações diretas quando houver incompatibilidade estrutural;
- explicitar todas as decisões de ajuste e agregação;
- priorizar consistência metodológica em detrimento de volume de dados.

Esses cuidados são essenciais para garantir a validade das conclusões.

# Métricas utilizadas

Para avaliar a desigualdade e a distribuição dos dados, serão utilizadas principalmente:

- totais por bairro;
- proporções relativas (ex: participação de cada bairro no total);
- distribuições estatísticas (mínimo, máximo, mediana e dispersão);
- comparações entre anos.

Sempre que necessário, os valores absolutos serão complementados por medidas relativas, para evitar interpretações distorcidas.
